#### <b> Indicador 1: Cantidad de aportantes Independientes, (tipo planilla igual a I) </b>

````python
def TipoAportante(x):
    Independiente = False
    if (x == 'I'):
        Independiente = True
    else:
        None
    return Independiente
````

````python
AportantesIndependientes = (DbMaestra.drop_duplicates(subset=['numero_identificacion'], keep='last')).loc[:, ['numero_identificacion', 'tipoplanilla']]
````

````python
print(AportantesIndependientes)
````

````python
AportantesIndependientes = (AportantesIndependientes['tipoplanilla']).apply(TipoAportante)
````

````python
print('La cantidad de aportantes imdependientes es:')
print(AportantesIndependientes[AportantesIndependientes == True].count())
````
