### <b> Importación y Lectura del Archivo CSV: Cotizantes </b> 


````python
cotizantes1 = pd.read_csv('Cotizantes.csv', sep=';', header=0, encoding='latin-1')
````

````python
cotizantes1.head(200)
````

````python
cotizantes2 = cotizantes1.drop_duplicates(subset=['cotizanteid'])
````

### <b> Estudio de Atributos del DataFrame: 'Cotizantes' </b>

````python
cotizantes1.info()
````

#### <b> Filtrado del DF de acuerdo al Periodo de Cotización </b>

````python
print(cotizantes2[(cotizantes2['periodo_cotizacion'] == 201901)])
````

````python
CotizantesFiltrado = cotizantes2[(cotizantes2['periodo_cotizacion'] == 201901)]
CotizantesFiltrado.head()
````

#### <b> Segregación de DF: Cotizantes, por campos; 'cotizanteid' y 'periodo_cotizacion' </b>

````python
print(cotizantes2[['cotizanteid','periodo_cotizacion']].groupby(['periodo_cotizacion']).count().sort_values(by='cotizanteid', ascending=False))
````

#### <b> Determinación la cantidad de personas que tiene pagos para el periodo de cotización=201901. </b>

````python
print(CotizantesFiltrado[['cotizanteid','periodo_cotizacion']].groupby(['periodo_cotizacion']).count().sort_values(by='cotizanteid', ascending=False))
````

#### <b> Determinación la cantidad de personas que tiene pagos por cada uno de los periodos de cotización. </b>

````python
print(cotizantes2[['cotizanteid','periodo_cotizacion']].groupby(['periodo_cotizacion']).count().sort_values(by='periodo_cotizacion', ascending=False))
````

#### <b> Calcular el promedio salarial de cada una de las personas. </b>

````python
print(cotizantes1[['cotizanteid','nmsalario_basico']].groupby(['cotizanteid']).mean().sort_values(by='nmsalario_basico', ascending=False))
````
